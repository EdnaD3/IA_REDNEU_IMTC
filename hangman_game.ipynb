```python
import random

def print_hangman(wrong):
    hangman_stages = [
        """
        +---+
        |   |
            |
            |
           ===
        """,
        """
        +---+
        |   |
        o   |
            |
           ===
        """,
        """
        +---+
        |   |
        o   |
        |   |
           ===
        """,
        """
        +---+
        |   |
        o   |
       /|   |
           ===
        """,
        """
        +---+
        |   |
        o   |
       /|\\  |
           ===
        """,
        """
        +---+
        |   |
        o   |
       /|\\  |
       /    ===
        """,
        """
        +---+
        |   |
        o   |
       /|\\  |
       / \\  ===
        """
    ]
    print(hangman_stages[wrong])

def printWord(randomWord, guessedLetters):
    result = ""
    for char in randomWord:
        if char in guessedLetters:
            result += char + " "
        else:
            result += "_ "
    return result

def main():
    print("Bienvenido :)")
    print("---------------------------------------------------------")

    palabras = ["rosa", "dorado", "cuervo", "pezglobo", "perro", "pato", "manzana", "pera", "uvas", "fresa"]

    # Elige una palabra al azar
    randomWord = random.choice(palabras)

    guessedLetters = []
    wrong_attempts = 0

    print("Adivina la palabra:")
    printWord(randomWord, guessedLetters)

    while wrong_attempts < 6:
        letterGuessed = input("\nIngresa una letra: ").lower()

        if len(letterGuessed) != 1 or not letterGuessed.isalpha():
            print("Ingresa una sola letra válida.")
            continue

        if letterGuessed in guessedLetters:
            print("Ya has adivinado esa letra.")
            continue

        guessedLetters.append(letterGuessed)

        if letterGuessed not in randomWord:
            wrong_attempts += 1
            print_hangman(wrong_attempts)
        else:
            printWordResult = printWord(randomWord, guessedLetters)
            print(printWordResult)

        if "_" not in printWordResult:
            print("¡Felicidades! Has adivinado la palabra.")
            break

    if "_" in printWordResult:
        print("Perdiste. La palabra era:", randomWord)

if __name__ == "__main__":
    main()
```
